```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
from tensorflow.keras.models import clone_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```python
# Нормалізація даних
x_train = x_train / 255
x_test = x_test / 255
```

```python
# Робимо категоріальну змінну
y_train = utils.to_categorical(y_train, 10)
y_test = utils.to_categorical(y_test, 10)

classes = ['футболка', 'брюки', 'светр', 'плаття', 'пальто',
'туфлі', 'сорочка', 'кросівки', 'сумка', 'черевики']
```

```python
# Створюємо послідовну модель
model = Sequential()

#Додаємо рівні мережі
model.add(Dense(800, input_dim=784, activation ="relu"))
model.add(Dense(10, activation = "softmax"))

#Компілюємо модель
model.compile(loss="categorical_crossentropy",
              optimizer="SGD",
              metrics=["accuracy"])

print(model.summary())

model.fit(x_train, y_train,
          batch_size=200,
          epochs=100,
          validation_split=0.2,
          verbose=1)

scores = model.evaluate ( x_test, y_test, verbose =1)
print ("Частка вірних відповідей на тестових даних, у відсотках:", round (scores[1]* 100, 4))
```

```python
# Обираємо випадковий індекс з тестової вибірки
random_idx = np.random.randint(0, x_test.shape[0])

# Беремо зображення та правильну мітку
image = x_test[random_idx]
true_label = np.argmax(y_test[random_idx])

# Робимо передбачення
prediction = model.predict(image.reshape(1, 784))  # Передбачаємо на одному зображенні
predicted_label = np.argmax(prediction)

# Виводимо результат
print(f"Передбачений клас: {predicted_label} ({classes[predicted_label]})")
print(f"Правильний клас: {true_label} ({classes[true_label]})")
```

```python
# Створення копії моделі, тільки з архітектурою, без навчання і ваг
model = clone_model(model)
model.compile(loss="categorical_crossentropy",
              optimizer="SGD",
              metrics=["accuracy"])

model.fit(x_train, y_train,
          batch_size=200,
          epochs=20,
          validation_split=0.2,
          verbose=1)

scores = model.evaluate ( x_test, y_test, verbose =1)
print ("Частка вірних відповідей на тестових даних, у відсотках:", round (scores[1]* 100, 4))

random_idx = np.random.randint(0, x_test.shape[0])

image = x_test[random_idx]
true_label = np.argmax(y_test[random_idx])

prediction = model.predict(image.reshape(1, 784))  # Передбачаємо на одному зображенні
predicted_label = np.argmax(prediction)

print(f"Передбачений клас: {predicted_label} ({classes[predicted_label]})")
print(f"Правильний клас: {true_label} ({classes[true_label]})")
```

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Завантаження та підготовка даних
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
x_train = x_train.reshape(60000, 784) / 255
x_test = x_test.reshape(10000, 784) / 255
y_train = utils.to_categorical(y_train, 10)
y_test = utils.to_categorical(y_test, 10)

# Функція для створення моделі (для KerasClassifier)
def create_model(optimizer='adam', units=128, dropout_rate=0.0):
    model = Sequential()
    model.add(Dense(units=units, input_dim=784, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(10, activation='softmax'))

    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=create_model, verbose=0)

# Параметри для перебору
param_grid = {
    'optimizer': ['adam', 'sgd'],  # Вибір оптимізатора
    'units': [64, 128, 256, 512],      # Кількість нейронів у прихованому шарі
    'dropout_rate': [0.0, 0.2, 0.5],  # Рівень Dropout
    'batch_size': [64, 128, 256, 512],  # Розмір пакета
    'epochs': [10, 20, 50, 100]             # Кількість епох
}

# Створюємо GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)

grid_result = grid.fit(x_train, y_train)

print(f"Найкращі гіперпараметри: {grid_result.best_params_}")

# Оцінка на тестовій вибірці з найкращою моделлю
test_accuracy = grid.score(x_test, y_test)
print(f"Точність на тестових даних: {test_accuracy:.4f}")
```
